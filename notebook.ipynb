```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpt
```

```python
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# load data
iris = datasets.load_iris()
print(iris)
```

```python
x = iris.data
y = iris.target
```

```python
print("Feature names: ",iris.feature_names)
print("Target names: ",iris.target_names)
print("Shape of data: ",x.shape)
```

```python
# split dataset into 80% train and 20% test
x_train, x_test , y_train , y_test = train_test_split(x , y , test_size =0.2, random_state = 42)
```

```python
def confusion_matrix_function(y_true , y_pred , model_name):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm,annot=True , fmt = "d" , cmap = "Blues", xticklabels= iris.target_names, yticklabels= iris.target_names)
    mpt.title(f"Confusion Matrix - {model_name}")
    mpt.xlabel("Predicted")
    mpt.ylabel("Actual")
    mpt.show()
```

```python
# 1 : Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
# train LR
log_reg = LogisticRegression(max_iter = 200)
log_reg.fit(x_train, y_train)
```

```python
# preditions
y_pred = log_reg.predict(x_test)
```

```python
# evaluation 
print("LR Accuracy",accuracy_score(y_test,y_pred))
print("\nClassification Report:\n", classification_report(y_test,y_pred))
```

```python
confusion_matrix_function(y_test,y_pred,"LR")
```

```python
# 2 : K-Nearest Neighbors(KNN)
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
yPred_knn = knn.predict(x_test)
```

```python
print("KNN accuracy: ", accuracy_score(y_test,yPred_knn))
print("\nClassification Report:\n", classification_report(y_test,yPred_knn))
```

```python
confusion_matrix_function(y_test, yPred_knn, "KNN")
```

```python
# 3 : Decision Tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state = 42)
```

```python
dt.fit(x_train,y_train)
```

```python
yPred_dt = dt.predict(x_test)
```

```python
print("Decision Tree Accuracy:" , accuracy_score(y_test, yPred_dt))
print("\nClassification Report:\n", classification_report(y_test,yPred_dt))
```

```python
confusion_matrix_function(y_test, yPred_dt, "DT")
```

```python
# 4 : Support Vector Machine(SVM)

from sklearn.svm import SVC
```

```python
svm = SVC(kernel="linear")
svm.fit(x_train,y_train)
```

```python
yPred_svm = svm.predict(x_test)
```

```python
print("SVM accuracty",accuracy_score(y_test , yPred_svm))
print("\nClassification Report:\n",classification_report(y_test, yPred_svm))
```

```python
confusion_matrix_function(y_test, yPred_svm, "SVM")
```

```python
#collection of Result
results = {
    "LR" : accuracy_score(y_test,y_pred),
    "KNN" : accuracy_score(y_test,yPred_knn),
    "DT" : accuracy_score(y_test,yPred_dt),
    "SVM" : accuracy_score(y_test,yPred_svm),       
           }
```

```python
for model , acc in results.items():
    print(f"{model} : {acc:.2f}")
```

```python
# Plot comparison

mpt.bar(results.keys(),results.values(), color = ['green',
    'pink' , 'blue' , 'yellow'])
mpt.ylabel("Accuracy")
mpt.title("Model Comparison on Iris Dataset")
mpt.show()
```
